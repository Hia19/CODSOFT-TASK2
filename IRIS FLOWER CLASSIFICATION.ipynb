### ***Applying Elbow Technique :***

```python
k_rng = range(1,10)
sse=[]
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(ir[['petal_length', 'petal_width']])
  sse.append(km.inertia_)
```

```python
km = KMeans(n_clusters = 3, random_state=0)
y_predicted = km.fit_predict(ir[['petal_length','petal_width']])
y_predicted
```
